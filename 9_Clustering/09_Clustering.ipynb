```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Dataset Information

titanic.info()
```

```python
# Missing Values
titanic.isnull().sum()
```

```python
# Statistical Summary

titanic.describe()
```

```python
# Survival Count Plot

plt.figure(figsize=(6,4))

sns.countplot(x='survived', data=titanic)

plt.title("Survival Count")

plt.xlabel("Survived")

plt.ylabel("Count")

plt.show()
```

```python
print(titanic.columns)
print(titanic[['sex', 'age', 'survived']].head())
```

```python
# Gender-wise Survival Plot

plt.figure(figsize=(6,4))

sns.countplot(x='sex',
              hue='survived',
              data=titanic)

plt.title("Survival Based on Gender")

plt.xlabel("Gender")

plt.ylabel("Count")

plt.show()
```

```python
# Passenger Class Plot

plt.figure(figsize=(6,4))

sns.countplot(x='pclass', data=titanic)

plt.title("Passenger Class Distribution")

plt.xlabel("Passenger Class")

plt.ylabel("Count")

plt.show()
```

### Observations:

1. The box plot shows the distribution of age for males and females based on survival.
2. Females have a higher survival rate compared to males.
3. Younger passengers (children) have higher chances of survival.
4. Median age of survivors is generally lower than non-survivors.
5. There are some outliers in age, especially among older passengers.
6. Male passengers show higher mortality compared to females.
