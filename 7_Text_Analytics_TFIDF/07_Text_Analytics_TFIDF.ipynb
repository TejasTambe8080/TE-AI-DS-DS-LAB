```python
# Import Required Libraries

import nltk
import pandas as pd

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

# Download Required Packages

```python
# Download Required Packages

nltk.download('punkt')

nltk.download('stopwords')

nltk.download('averaged_perceptron_tagger')

nltk.download('wordnet')
```

# Step 1: Create Sample Document

```python
# Sample Document

document = """
Artificial Intelligence is transforming the world.
Machine learning and Natural Language Processing are important fields of AI.
AI helps in automation, prediction, and data analysis.
"""

print(document)
```

# Step 2: Tokenization

Tokenization splits the document into individual words.

```python
# Tokenization

tokens = word_tokenize(document)

print("Tokens:")
print(tokens)
```

# Step 3: POS Tagging

POS Tagging identifies grammatical parts of speech.

```python
# POS Tagging

pos_tags = pos_tag(tokens)

print("POS Tags:")
print(pos_tags)
```

# Step 4: Stop Words Removal

Stop words are common words that do not add important meaning.

```python
# Remove Stop Words

stop_words = set(stopwords.words('english'))

filtered_words = [
    word for word in tokens
    if word.lower() not in stop_words
]

print("Words after Stop Words Removal:")
print(filtered_words)
```

# Step 5: Stemming

Stemming converts words into root form.

```python
# Apply Stemming

stemmer = PorterStemmer()

stemmed_words = [
    stemmer.stem(word)
    for word in filtered_words
]

print("Stemmed Words:")
print(stemmed_words)
```

# Step 6: Lemmatization

Lemmatization converts words into meaningful base form.

```python
# Apply Lemmatization

lemmatizer = WordNetLemmatizer()

lemmatized_words = [
    lemmatizer.lemmatize(word)
    for word in filtered_words
]

print("Lemmatized Words:")
print(lemmatized_words)
```

# Step 7: Create Term Frequency (TF) Representation

```python
# Create TF Representation

documents = [document]

count_vectorizer = CountVectorizer()

tf_matrix = count_vectorizer.fit_transform(documents)

tf_df = pd.DataFrame(
    tf_matrix.toarray(),
    columns=count_vectorizer.get_feature_names_out()
)

print("Term Frequency Matrix:")
print(tf_df)
```

# Step 8: Create TF-IDF Representation

TF-IDF gives importance score to words.

Formula:

TF-IDF = TF × IDF

```python
# Create TF-IDF Representation

tfidf_vectorizer = TfidfVectorizer()

tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out()
)

print("TF-IDF Matrix:")
print(tfidf_df)
```
